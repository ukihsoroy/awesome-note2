## Matplot Scatter 散点图的绘制

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
# 绘制散点图
plt.plot(x, np.sin(x), 'o')
# plt.scatter(x, np.sin(x))
plt.show()
```

> 从上面的示例可以看到，使用plot和scatter绘制散点图是没有去别的
> 但是使用plot绘制图形的速度优于scatter，所以如果只是画一堆点并且点的形式没有差别，那么我们使用plot。如果点的形式有差别（大小和颜色），那么我们必须使用scatter。

```python
np.random.seed(0)
x = np.random.rand(100)
y = np.random.rand(100)
```

```python
# point(x)
# 绘制散点图
# 生成10种大小
size = np.random.rand(10) * 1000
color = np.random.rand(100)
# 注意点的个数和点的大小可以不同
# 点的颜色和点的个数必须不同
plt.scatter(x, y, s=size, c=color, alpha=0.7)
plt.show()
```

### 绘制不同颜色的线条

```python
x = np.linspace(0, 10, 100)
plt.plot(x, x+0, '--g')
plt.plot(x, x+1, '-.r')
plt.show()
```
