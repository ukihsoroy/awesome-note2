```python
import matplotlib.pyplot as plt
```

```python
# 绘制直线, x,y的坐标
plt.plot([0, 2], [1, 4])
plt.show()
```

```python
# 绘制折线
X = [1, 2, 3, 4, 5]
Y = [1, 4, 9, 16, 25]
plt.plot(X, Y)
plt.show()
```

## 绘制一元二次方程

```python
x = range(-100, 100)
y = [i ** 2 for i in x]
plt.plot(x, y)
plt.show()
```

## 绘制正弦余弦曲线

```python
import numpy as np
```

```python
## 生成0-10之间 100个等差数
x = np.linspace(0, 10, 100)
sin_y = np.sin(x)
```

```python
plt.plot(x, sin_y)
plt.show()
```

### 绘制余弦

```python
cos_y = np.cos(x)
```

```python
plt.plot(x, cos_y)
plt.plot(x, sin_y)
plt.show()
```

## subplot的使用

```python
# 对画布进行分区
plt.subplot(2, 2, 1)
# 修改x, y轴的坐标
plt.xlim(-5, 20)
plt.ylim(-2, 2)
plt.plot(x, sin_y)

plt.subplot(2, 2, 3)
plt.plot(x, cos_y)
plt.show()
```
