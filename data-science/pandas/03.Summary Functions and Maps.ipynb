## Maps

```python
review_lead_time_mean = reviews.lead_time.mean()
reviews.lead_time.map(lambda p: p - review_lead_time_mean)
```

```python
def remean_lead_time(row):
    row.lead_time = row.lead_time - review_lead_time_mean
    return row

reviews.apply(remean_lead_time, axis='columns')
```
